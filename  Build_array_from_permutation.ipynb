We want to create a list ans from a list nums where ans[i] = nums[nums[i]] :

```python
def buildArray(nums):
    return [nums[num] for num in nums]
```

Example 1:

```python
nums = [0,2,1,5,3,4]
#expected result: [0,1,2,4,5,3]
buildArray(nums)
```

Example 2:

```python
nums = [5,0,1,2,3,4]
#expected result: [4,5,0,1,2,3]
buildArray(nums)
```
